# Exploratory Data Analysis


```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
#Folder path must end with /
FOLDER = "/Users/echo/Documents/0424_Data_Enrichment/Project-3/Data/"
```

```python
#Reading in 2000 movies from above output
results00 = pd.read_csv("/Users/echo/Documents/0424_Data_Enrichment/Project-3/Data/final_tmdb_data_2000.csv.gz")
results00
```

```python
#Reading in 2001 movies from above output
results01 = pd.read_csv("/Users/echo/Documents/0424_Data_Enrichment/Project-3/Data/final_tmdb_data_2001.csv.gz")
results01
```

```python
#Concatenate the data into 1 dataframe for the remainder of the analysis.
df = pd.concat([results00, results01], ignore_index=True)
df.info()
```

1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.

```python
#Filter for movies that made money.
valid_movies = df[(df["budget"] > 0) | (df["revenue"] > 0)]
valid_movies = valid_movies.shape[0]
valid_movies
```

2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
#Finding count for each certification category
ratings = df["certification"].value_counts()
```

```python
fig, ax = plt.subplots()
sns.histplot(df["certification"])
ax.set_title("Number of Movies per Rating", weight="bold")
ax.set_xlabel("Ratings", weight="bold")
ax.set_ylabel("Number of Movies", weight="bold");
```

3. What is the average revenue per certification category? (per million)

```python
#Grouping ratings together and finding average revenue
avg_revenue_per_rating = df.groupby("certification")["revenue"].mean()
avg_revenue_per_rating
```

```python
fig, ax = plt.subplots()
sns.barplot(df, x=df["certification"], y=df["revenue"])
ax.set_title("Revenue per Rating", weight="bold")
ax.set_xlabel("Rating", weight="bold")
ax.set_ylabel("Revnue", weight="bold");
```

4. What is the average budget per certification category? (per million)

```python
#Getting the budget of each rating
avg_budget_per_rating = df.groupby("certification")["budget"].mean()
avg_budget_per_rating
```

```python
fig, ax = plt.subplots()
sns.barplot(df, x=df["certification"], y=df["budget"])
ax.set_title("Budget per Rating", weight="bold")
ax.set_xlabel("Ratings", weight="bold")
ax.set_ylabel("Budget", weight="bold");
```

```python
#Saving combined dataset into one file
df.to_csv(f"{FOLDER}tmdb_results_combined.csv.gz", compression="gzip", index=False)
```
